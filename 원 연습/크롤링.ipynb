```python
from datetime import date, timedelta
import pandas as pd
import time

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date='
YY = int(input("기준 년도를 입력하세요(예:2020):"))
mm = int(input("기준 월을 입력하세요(예:7):"))
dd = int(input("기준 일을 입력하세요(예:7):"))
period = int(input("크롤링을 하고자 하는 기간을 입려해 주세요:"))
start_date = date(YY,mm,dd)
timegap = timedelta(days=period)
start_date = start_date - timegap
start_date = start_date.strftime('%Y-%m-%d')
range_date = pd.date_range(start_date, periods=period, freq='D')

tit_lst=[]
point_lst=[]
date_lst=[]
for today in range_date:
    html = requests.get(url+today.strftime('%Y%m%d'))
    time.sleep(0.1)
    soup = BeautifulSoup(html.text, 'html.parser')
    
    tit5_tag = soup.find_all('div',class_='tit5')
    point_tag = soup.find_all('td',class_='point')
    for num in range(len(tit5_tag)):
        title = tit5_tag[num].find('a').text
        tit_lst.append(title)
        point=point_tag[num].text
        point_lst.append(point)
        date_lst.append(today.strftime('%Y-%m-%d'))
        
movie_df = pd.DataFrame({'date':date_lst,'name':tit_lst,'point':point_lst})
movie_df.to_csv("c:/py_data/네이버 영화 평점 크롤링.csv", encoding='utf-8')
```

```python
url = 'https://music.bugs.co.kr/chart/track/day/total?chartdate='
YY = int(input("기준 년도를 입력하세요(예:2020):"))
mm = int(input("기준 월을 입력하세요(예:7):"))
dd = int(input("기준 일을 입력하세요(예:7):"))
period = int(input("크롤링을 하고자 하는 기간을 입려해 주세요:"))
start_date = date(YY,mm,dd)
timegap = timedelta(days=period)
start_date = start_date - timegap
start_date = start_date.strftime('%Y-%m-%d')
range_date = pd.date_range(start_date, periods=period, freq='D')
title_lst=[]
rank_lst=[]
date_lst=[]

for today in range_date:
    html = requests.get(url+today.strftime('%Y%m%d'))
    time.sleep(0.1)
    soup = BeautifulSoup(html.text, 'html.parser')
    
    title_tag = soup.find_all('p',class_='title')
    rank_tag = soup.find_all('div',class_='ranking')
    for num in range(len(title_tag)):
        title = title_tag[num].find('a').text
        title_lst.append(title)
        rank = rank_tag[num].find('strong').text
        rank_lst.append(rank + '위')
        date_lst.append(today.strftime('%Y-%m-%d'))
        
movie_df = pd.DataFrame({'date':date_lst,'title':title_lst,'ranking':rank_lst})
movie_df.to_csv("c:/py_data/벅스뮤직 차트 크롤링.csv", encoding='utf-8')
```
